## Exercise 2  
1. Построить график функции $sin(x)/x$ на отрезке $[-6\pi;6\pi]$ и добавить тики по оси $x$ с шагом $\pi/2$ от $-6\pi$ до $6\pi$, причём обозначения тиков должны содержать греческую букву $\pi$. Добавить сетку на график.

```python

pi = math.pi
x = np.linspace(-6 * pi, 6 * pi , 100)
y = np.sin(x)/x

x_pi = np.arange(-6 * pi, 6 * pi+1, pi/2)
y_pi = np.sin(x_pi)/x_pi


fig, ax = plt.subplots(figsize=(15,4))

ax.plot(x, y, color='pink')
ax.plot(x_pi, y_pi,  color='black', linestyle='', marker='$\pi$')
ax.grid()



ax.set_xticks(np.arange(-6 * pi, 6 * pi + 1, pi/2))
pii = [(str(-12+i) + 'π/2' if i != 12 else '0') for i in range(25)]
leng = len(np.arange(-6 * pi, 6 * pi, pi/2))

#pii = (np.arange(-6 * pi, 6 * pi, pi/2))
#print(pii)
ax.set_xticklabels(pii)
print()


```

## Exercise 3  
1. Создайте цветовой график $sin(x)sin(y)$ на интервале $[−π,π]×[−π,π]$.
2. Создайте цветовой график $e^{-x^2-y^2}$ на интервале $[−3,3]×[−3,3]$.

```python
x = np.arange(-pi, pi, 0.1)
y = x
X, Y = np.meshgrid(x,y)
Z = np.sin(X) * np.sin(Y)

fig, ax = plt.subplots(figsize=(6,5))

norm = mpl.colors.Normalize(-pi, pi)

p = ax.pcolor(X, Y, Z, norm = norm, cmap=mpl.cm.summer)
ax.axis()

ax.set_title('1')
cb = fig.colorbar(p, ax=ax)
```

```python
x = np.arange(-3, 3, 0.1)
y = x
X, Y = np.meshgrid(x,y)
Z = math.e**(-X**2 - Y**2)

fig, ax = plt.subplots(figsize=(6,5))

norm = mpl.colors.Normalize(-pi, pi)

p = ax.pcolor(X, Y, Z, norm = norm, cmap=mpl.cm.bwr)
ax.axis()

ax.set_title('2')
cb = fig.colorbar(p, ax=ax)
```
